```
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```
x,y=make_blobs(
    n_samples=200,
    n_features=2,
    centers=3,
    cluster_std=1,
    random_state=0
)
```

```
plt.figure(figsize=(14,8))
plt.scatter(
    x[:,0],x[:,1],
    c='white',
    edgecolor='black'
)
plt.grid()
plt.show()
```

```
kmeans=KMeans(
    n_clusters=3,
    init='random',
    max_iter=4,
    random_state=2
)
```

```
y_km=kmeans.fit_predict(x)
y_km
```

```
plt.figure(figsize=(14,8))
plt.scatter(
    x[y_km==0,0],x[y_km==0,1],
    s=50,c='lightgreen',
    marker='s',edgecolor='black',
    label='cluster 1'
)
plt.scatter(
    x[y_km==1,0],x[y_km==1,1],
    s=50,c='yellow',
    marker='o',edgecolor='black',
    label='cluster 2'
)
plt.scatter(
    x[y_km==2,0],x[y_km==2,1],
    s=50,c='pink',
    marker='v',edgecolor='black',
    label='cluster 3'
)
plt.scatter(
    kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],
    s=250,marker='*',
    c='red',
    edgecolor='black',
    label='centeroids'
)
plt.legend(scatterpoints=1)
plt.show()
```
